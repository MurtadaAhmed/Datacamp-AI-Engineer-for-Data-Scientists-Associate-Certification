```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd

dataset = "datasets/company-stock-movements-2010-2015-incl.csv"
df_company_stocks = pd.read_csv(dataset)

companies = df_company_stocks.iloc[0:, 0]
movements = df_company_stocks.iloc[0:, 1:]

model = TSNE(learning_rate=50)

tsne_features = model.fit_transform(movements)

xs = tsne_features[:, 0]
ys = tsne_features[:, 1]

plt.scatter(xs, ys, alpha=0.5)

for x, y, company in zip(xs, ys, companies):
    plt.annotate(company, (x, y), fontsize=5, alpha=0.75)

plt.show()
```
