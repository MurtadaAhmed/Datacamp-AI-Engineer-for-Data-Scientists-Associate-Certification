# Transforming features for better clusterings

```python
from sklearn.cluster import KMeans
dataset = "datasets/wine.csv"
df_wine = pd.read_csv(dataset)
df_wine_labels = df_wine.iloc[1:, 2:] # extracting the data without the first row and the first two columns
df_wine_varieties = df_wine.iloc[1:, 1] # extracting the varieties from the second column, excluding the first row

model = KMeans(n_clusters=3)
labels = model.fit_predict(df_wine_labels)

df = pd.DataFrame({"labels": labels, "varieties": df_wine_varieties})
ct = pd.crosstab(df['labels'], df['varieties'])

ct
```

```python
# using pipeline to combine StandardScaler and KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
kmeans = KMeans(n_clusters=3)

pipeline = make_pipeline(scaler, kmeans)

pipeline.fit(df_wine_labels)

labels = pipeline.predict(df_wine_labels)

df = pd.DataFrame({"labels": labels, "varieties": df_wine_varieties})
ct = pd.crosstab(df['labels'], df['varieties'])

ct
```
