# Slicing and subsetting with .loc and .iloc

## loc / iloc

```python
# we must have already used sort_index on the dataframe
homelessness_df
homelessness_df = homelessness_df.set_index(["state", "region"]) # setting state as outer index and region as inner index
homelessness_df = homelessness_df.sort_index() # sorting the index
homelessness_df
```

```python
# slicing at the outer level of the index
homelessness_df.loc["Alaska":"Colorado"]
```

```python
# to slice including the inner index, we need to pass the first and last positions as tubles
homelessness_df.loc[("Alaska", "Pacific"):("California", "Pacific")]
```

```python
# we can also slice columns (here we mention the column names that are not included in the index)
homelessness_df.loc[:, "individuals": "family_members"] # we don't mention "state" and "region" as they are included in the index
```

```python
# slicing rows and columns at the same time
homelessness_df.loc["Alaska": "California", "family_members":"state_pop"]
```

```python
# iloc to slice by rows and columns numbers
homelessness_df.iloc[1:3, 1:2] # using index, the last index is excluded
```
